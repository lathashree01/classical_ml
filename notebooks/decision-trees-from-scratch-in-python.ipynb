# Decision Trees and Random Forests from scartch in Python

A decision tree is a non-parametric supervised learning algorithm used for both classification and regression tasks. It features a hierarchical tree structure that includes a root node, branches, internal nodes, and leaf nodes.

## Algorithm


* **Feature Selection**: The algorithm selects the best attribute to split the data using measures like Gini impurity, information gain, or entropy.
* **Splitting**: The dataset is divided into smaller subsets based on the selected attribute with highest score.
* **Recursive Partitioning**: This process is repeated recursively for each child node until a stopping criterion is met where a stopping condition could be maximum depth of tree or number of iterations.


* *To prevent overfitting*, the tree could be pruned by removing branches that provide little classification power to the entire decision tree.
